```python
#pakage
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#한글 처리
plt.rcParams["font.family"] = 'GULIM'
plt.rcParams["figure.figsize"] = (10,10)

#파일 부르기
file = "ETO_13_20201207_G_001.shp"
data = gpd.read_file(file, encoding = 'utf-8')
data.head(50)
```

```python
#plot
ax1 = data.convex_hull.plot(color='red', edgecolor="black")
ax1.set_title("전국")
ax1.set_axis_off() # 축 범례 삭제
plt.show(ax1)
```

```python
# 대한민국 상세주소 지도 (읍면동리까지)
plt.rcParams["font.family"] = 'GULIM'
plt.rcParams["figure.figsize"] = (100,100)

#
file2 = "TL_SCCO_LI.shp"
data2 = gpd.read_file(file2, encoding = 'euc-kr')

#
ax2 = data2.convex_hull.plot(color='lightgreen', edgecolor="black")
ax2.set_title("전국")
ax2.set_axis_off() # 축 범례 삭제 (안이쁘니까~)
plt.show(ax2)
```
